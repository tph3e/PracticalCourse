This file primarily includes sanity checks and general comparisons between the given and the simulated log.

```python
import pandas as pd
import pm4py
```

```python
LOG1_PATH = "../data/BPI Challenge 2017.xes"
LOG2_PATH = "../data/generated_log.xes"

CASE_COL = "case:concept:name"
ACTIVITY_COL = "concept:name"
TIME_COL = "time:timestamp"
```

```python
def load_log(path):
    df = pm4py.read_xes(path, variant="r4pm")

    df[TIME_COL] = pd.to_datetime(df[TIME_COL])

    df = df.sort_values([CASE_COL, TIME_COL])

    return df

log1 = load_log(LOG1_PATH)
log2 = load_log(LOG2_PATH)
```

```python
print(log2[ACTIVITY_COL].unique())
print(log1[ACTIVITY_COL].unique())
```
